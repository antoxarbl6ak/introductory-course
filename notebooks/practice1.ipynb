```python
import pandas as pd

df = pd.read_csv(r"..\data\french-patents.csv")
out = open(r"..\data\practice1.txt", 'w')
```

```python
df["number_words"] = df["patent"].apply(lambda x: len(x.replace('.', '').replace(',', '').lower().split()))
df
```

```python
out.write("количественный анализ текста патента\n\n")
out.write(df["number_words"].describe().to_string())
out.write(f"\n\nколичество уникальных слов: {df['patent'].apply(lambda x: len(set(x.replace('.', '').replace(',', '').lower().split()))).sum()}")
```

```python
import datetime as dt

print("\n\nдиапозон дат:\n\n", file=out)
print(df["date of publication"].apply(lambda x: dt.date(*list(map(int, x[-10:].split('.')))[::-1])).agg(["min", "max"]), file=out)
```

```python
out.close()
```
